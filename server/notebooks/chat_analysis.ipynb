# chat.db 분석 노트북

chat_logs 테이블 구조:
- `id`, `channel_id`, `room_name`, `user_hash`, `user_name`, `content`, `log_id`, `timestamp`, `created_at`, `source`
- 약 167만 행

```python
import sqlite3
import pandas as pd
from datetime import datetime

DB_PATH = "../data/chat.db"
conn = sqlite3.connect(DB_PATH)
print("DB 연결 완료")
```

```python
# 전체 행 수 확인
total = pd.read_sql_query("SELECT COUNT(*) as cnt FROM chat_logs", conn)
print(f"전체 채팅 수: {total['cnt'][0]:,}")
```

```python
# 방별 채팅 수
pd.read_sql_query("""
    SELECT room_name, COUNT(*) as cnt 
    FROM chat_logs 
    GROUP BY room_name 
    ORDER BY cnt DESC
""", conn)
```

```python
# 유저별 채팅 수 (상위 20명)
pd.read_sql_query("""
    SELECT user_name, COUNT(*) as cnt 
    FROM chat_logs 
    GROUP BY user_hash 
    ORDER BY cnt DESC 
    LIMIT 20
""", conn)
```

```python
# 최근 100건 미리보기
df_recent = pd.read_sql_query("""
    SELECT * FROM chat_logs 
    ORDER BY id DESC 
    LIMIT 100
""", conn)
df_recent.head(10)
```
