```python
import requests
import json # JSON을 예쁘게 출력하기 위해 import

server_name = "칼릭스" 
user_id = "로스트아크하실분"
url = f"https://mobingi.ngrok.io/api2/rank/{server_name}/{user_id}"

print(f"요청 URL: {url}\n")

try:
    response = requests.get(url)
    response.raise_for_status() 
    print("✅ 요청 성공!")
    print("--- null 값인 키-값 쌍을 제외하고 출력 ---")

    # 1. response.text 대신 response.json()으로 데이터를 파싱합니다.
    #    서버 응답이 [{}, {}, {}] 형태의 리스트라고 가정합니다.
    data_list = response.json()

    # 2. 리스트의 각 아이템(객체)을 순회합니다.
    for item in data_list:
        # 3. 딕셔너리 컴프리헨션을 사용해 값이 None(json의 null)이 아닌
        #    키-값 쌍만 골라서 새로운 딕셔너리를 만듭니다.
        cleaned_item = {
            key: value 
            for key, value in item.items() 
            if value is not None
        }
        
        # 4. 정리된 객체를 JSON 문자열로 예쁘게 출력
        print(json.dumps(cleaned_item, indent=2, ensure_ascii=False))
        print("-" * 20) # 객체별 구분선

except requests.exceptions.HTTPError as http_err:
    print(f"❌ HTTP 오류 발생: {http_err}")
except requests.exceptions.RequestException as err:
    print(f"❌ 요청 중 기타 오류 발생: {err}")
except requests.exceptions.JSONDecodeError:
    # 만약 서버가 [{}, {}] 형태가 아닌, {} {} {} 처럼
    # 잘못된 JSON을 보낸다면 여기서 오류가 날 수 있습니다.
    # 이 경우, response.text를 다른 방식으로 처리해야 합니다.
    print("❌ JSON 파싱 오류: 서버 응답이 유효한 JSON 형식이 아닙니다.")
    print(f"서버 원본 응답: {response.text}")
```

```python
import requests
import json

# 1. 실제 값으로 설정
server_name = "칼릭스" 
user_id = "크앙박"
url = f"https://mobingi.ngrok.io/api2/rank/{server_name}/{user_id}"

print(f"요청 URL: {url}\n")

try:
    response = requests.get(url)
    # 요청 실패 시 오류 발생 (예: 404, 500)
    response.raise_for_status() 
    print("✅ 요청 성공!")

    # 1. 서버 응답을 JSON 리스트로 파싱합니다.
    data_list = response.json()

    # 2. 리스트가 비어있는지 확인합니다.
    if not data_list:
        print("❌ 응답 데이터가 비어있습니다.")
    else:
        # 3. 'level' 값을 기준으로 가장 높은 객체를 찾습니다.
        #    max() 함수와 key=lambda를 사용합니다.
        #    .get("level", 0)은 혹시 'level' 키가 없더라도 
        #    오류 없이 0으로 간주하여 비교하도록 합니다.
        main_class = max(data_list, key=lambda item: item.get("level", 0))

        # 4. 찾은 메인 클래스 객체에서 null(None) 값인 키-값 쌍을 제거합니다.
        cleaned_main_class = {
            key: value 
            for key, value in main_class.items() 
            if value is not None
        }
        
        print("--- 🏆 'level'이 가장 높은 클래스 정보 ---")
        
        # 5. 최종 결과를 예쁘게 출력합니다.
        #    indent=2 : 들여쓰기
        #    ensure_ascii=False : 한글이 깨지지 않게 출력
        print(json.dumps(cleaned_main_class, indent=2, ensure_ascii=False))

except requests.exceptions.HTTPError as http_err:
    # 404, 500 등 서버에서 오류 코드를 보낸 경우
    print(f"❌ HTTP 오류 발생: {http_err}")
    print(f"서버 응답: {response.text}")
except requests.exceptions.RequestException as err:
    # 네트워크 연결 실패 등 요청 자체에 문제가 생긴 경우
    print(f"❌ 요청 중 기타 오류 발생: {err}")
except requests.exceptions.JSONDecodeError:
    # 서버 응답이 유효한 JSON 형식이 아닌 경우
    print("❌ JSON 파싱 오류: 서버 응답이 유효한 JSON 형식이 아닙니다.")
    print(f"서버 원본 응답: {response.text}")
```

```python
import requests
import json

# API 기본 URL
base_url = "https://mobingi.ngrok.io"
endpoint = "/api/item/rune"
url = base_url + endpoint

# --- 💡 테스트 케이스 선택 ---

# Case 1: 파라미터 없이 전체 조회 (테스트 시 아래 Case 2를 주석 처리)
# params = {} 
# print(f"요청 URL: {url} (전체 룬 조회)")

# Case 2: 특정 조건으로 검색 (예: 이름, 등급 등)
# (API 문서에 있는 파라미터를 자유롭게 추가/제거하세요)
params = {
    "label": "연쇄",  # 검색할 룬 이름 (예: "축복의 룬")
    
    # "category": "예시_카테고리",
    "star": 8
}
print(f"요청 URL: {url}")
print(f"파라미터: {params}\n")


# --------------------------

try:
    # .get() 요청 시 params 인자로 딕셔너리를 전달
    response = requests.get(url, params=params)
    response.raise_for_status() 
    print("✅ 요청 성공!")

    data = response.json()

    print("--- 룬 정보 (null 값 제외) ---")

    # 응답이 리스트일 경우 (아마도)
    if isinstance(data, list):
        for item in data:
            cleaned_item = {k: v for k, v in item.items() if v is not None}
            print(json.dumps(cleaned_item, indent=2, ensure_ascii=False))
            print("-" * 20)
    # 응답이 단일 객체일 경우
    elif isinstance(data, dict):
        cleaned_item = {k: v for k, v in data.items() if v is not None}
        print(json.dumps(cleaned_item, indent=2, ensure_ascii=False))
    # 기타
    else:
        print(data)

except requests.exceptions.HTTPError as http_err:
    print(f"❌ HTTP 오류 발생: {http_err}")
    print(f"서버 응답: {response.text}")
except requests.exceptions.RequestException as err:
    print(f"❌ 요청 중 기타 오류 발생: {err}")
except requests.exceptions.JSONDecodeError:
    print(f"❌ JSON 파싱 오류: {response.text}")
```

```python
import requests
import json
import pandas as pd
from datetime import datetime

def get_user_detail(server, nickname):
    """API로부터 유저 상세 정보를 가져오는 함수"""
    base_url = "https://mobingi.ngrok.io"
    url = f"{base_url}/api2/detail/{server}/{nickname}"
    
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"❌ 상세 정보 조회 중 오류 발생: {e}")
        return None

def display_user_summary(data):
    """데이터를 가공하여 최종 결과물(마지막 코드 기준)을 출력하는 함수"""
    if not data or 'rankList' not in data:
        print("검색 결과 데이터가 없거나 올바르지 않습니다.")
        return

    # 1. 메인 클래스 특정 (enable_class == "1")
    main_class = next((item for item in data['rankList'] if item.get('enable_class') == "1"), None)
    
    if not main_class:
        # 만약 메인 클래스 설정이 없다면 첫 번째 항목을 기본값으로 사용
        main_class = data['rankList'][0]
        is_main_found = False
    else:
        is_main_found = True

    # 2. 날짜 포맷 변경 (ISO 8601 -> YYYY-MM-DD HH:MM)
    raw_date = main_class.get('update_at')
    if raw_date:
        try:
            formatted_date = datetime.strptime(raw_date, "%Y-%m-%dT%H:%M:%S.%fZ").strftime("%Y-%m-%d %H:%M")
        except ValueError:
            formatted_date = raw_date 
    else:
        formatted_date = "정보 없음"

    # 3. 메인 클래스의 히스토리 필터링 및 변화량 계산
    main_type = main_class.get('class_type')
    history_list = [h for h in data.get('history', []) if h.get('class_type') == main_type]
    
    # 변화량 계산을 위해 최근 8개 추출
    recent_8_days = history_list[-8:]
    display_history = []
    
    for i in range(len(recent_8_days) - 1, 0, -1):
        curr = recent_8_days[i]
        prev = recent_8_days[i-1]
        
        diff = curr['level'] - prev['level']
        diff_str = f"(+{diff:,})" if diff > 0 else f"({diff:,})" if diff < 0 else "(+0)"
        
        d = curr['date_time']
        date_str = f"20{d[:2]}-{d[2:4]}-{d[4:]}"
        
        display_history.append(f"[{date_str}] 전투력: {curr['level']:,} {diff_str}")

    # --- 출력부 (사용자 요청 포맷) ---
    print(f"닉네임: {main_class['user_id']}")
    print(f"서버: {main_class['server_name']}")
    print(f"직업: {main_class['class_name']}{' (메인)' if is_main_found else ''}")
    print(f"서버내 순위: {main_class.get('server_rank', main_class.get('rank', '-'))}")
    print(f"직업 순위: {main_class.get('class_rank', '-')}")
    print("")
    print(f"종합 평가: {main_class.get('evaluation_score', 0):,}")
    print(f"전투력: {main_class.get('level', 0):,}")
    print(f"생활력: {main_class.get('attractiveness', 0):,}")  # attractiveness -> 생활력임
    print(f"매력: {main_class.get('vitality', 0):,}") # vitality -> 매력임
    print("")
    print(f"마지막 접속: {formatted_date}")
    print("")
    print("최근 성장 추이 (최신순 7일)")
    print("-" * 40)
    
    if not display_history:
        print("최근 성장 기록이 없습니다.")
    else:
        for line in display_history[:7]:
            print(line)
    print("-" * 40)

# --- 메인 실행부 ---
if __name__ == "__main__":
    # 1. 설정
    server_name = "칼릭스"
    user_nickname = "히로롱"
    
    # 2. 데이터 가져오기
    detail_info = get_user_detail(server_name, user_nickname)
    
    # 3. 화면 출력
    if detail_info:
        display_user_summary(detail_info)
```
